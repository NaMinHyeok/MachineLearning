## np.rand와 np.randn

```python
a = np.random.rand(500000)
b = np.random.randn(500000)

plt.hist(a, density=True, bins=100, histtype="step", color="blue", label="rand")
plt.hist(b, density=True, bins=100, histtype="step", color="red", label="randn")
plt.axis([-5.5, 5.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

## 산술 연산

## 일반 함수
넘파이는 일반 함수(universal function) 또는 ufunc라고 부르는 원소별 함수를 제공합니다. 예를 들면 square 함수는 원본 ndarray를 복사하여 각 원소를 제곱한 새로운 ndarray 객체를 반환합니다:

```python
a = np.array([[-1.5, 4.1, 5], [15, 20, 25]])
np.square(a)
```

```python
print("원본 ndarray")
print(a)
for func in (np.abs, np.sqrt, np.exp, np.log, np.sign, np.ceil, np.modf, np.isnan, np.cos):
    print("\n", func.__name__)
    print(func(a))
```

# 배열 쌓기
종종 다른 배열을 쌓아야 할 때가 있습니다. 넘파이는 이를 위해 몇 개의 함수를 제공합니다. 먼저 배열 몇 개를 만들어 보죠.

```python
q1 = np.full((4,4), 1.0)
q1
```

```python
q2 = np.full((5,4), 2.0)
q2
```

```python
q3 = np.full((4,4), 3.0)
q3
```

## vstack
vstack 함수를 사용하여 수직으로 쌓아보자

```python
q4 = np.vstack((q1, q2, q3))
q4
```

```python
q4.shape
```

##hstack
hstack을 사용해 수평으로도 쌓을 수 있다.

```python
q5 = np.hstack((q1, q3))
q5
```

```python
q5.shape
```

q1 q3는 모두 4개의 행 하지만 q2는 5개의 행
따라서 q2는 q1 q3와 수평으로 쌓을 수 없다.

```python
try:
    q5 = np.hstack((q1, q2, q3))
except ValueError as e:
    print(e)
```

##concatenate
concatenate 함수는 지정한 축으로도 배열을 쌓는다.

```python
q7 = np.concatenate((q1, q2, q3), axis=0)  # vstack과 동일
q7
```

```python
q7.shape
```

```python
q8 = np.stack((q1, q3))
q8
```

## 특잇값 분해

```python
U, S_diag, V = linalg.svd(m4)
U
```

```python
S_diag
```

```python
S = np.zeros((4,5))
S[np.diag_indices(4)] = S_diag
S
```

```python
V
```

```python
U.dot(S).dot(V)
```

##선형 방정식 풀기

solve 함수는 다음과 같은 선형 방정식을 풉니다.

4x+6y = 6

10x + 3y = -9

```python
coeffs = np.array([[4,6],[10,3]])
depvars = np.array([6,-9])
solution = linalg.solve(coeffs, depvars)
solution
```

```python
coeffs.dot(solution), depvars
```

```python
np.allclose(coeffs.dot(solution),depvars)
```

# 벡터화

한 번에 하나씩 개별 배열 원소에 대해 연산을 실행하는 대신 배열 연산을 사용하면 훨씬 효율적인 코드를 만들 수 있습니다. 이를 벡터화라고 합니다. 이를 사용하여 넘파이의 최적화된 성능을 활용할 수 있습니다.

```python
import math
data = np.empty((768,1024))
for y in range(768):
  for x in range(1024):
    data[y,x] = math.sin(x*y/40.5)
```

```python
x_coords = np.arange(0,1024)
y_coords = np.arange(0, 768)
X, Y = np.meshgrid(x_coords, y_coords)
X
```

```python
Y
```

```python
data = np.sin(X*Y/40.5)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
fig = plt.figure(1, figsize = (7,6))
plt.imshow(data, cmap = cm.hot)
plt.show()
```
